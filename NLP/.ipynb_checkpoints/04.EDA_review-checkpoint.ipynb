###### 데이터 분석
1. 데이터 크기
2. 데이터 개수
3. 각 리뷰의 문자 길이분호
4. 많이 사용된 단어
5. 긍정, 부정 데이터의 분포

7. 특수문자 및 대문자, 소문자 비율

```python
print("파일 크기: ")
for file in os.listdir('data_in/'):
    if 'tsv' in file and 'zip' not in file:
        print(file.ljust(30) + str(round(os.path.getsize('data_in/' + file) / 1000000,2)) + 'MB')
```

```python
print('전체 학습 데이터의 개수: {}'.format(len(train_data)))
```

```python
# 'review'열의 각 리뷰의 데이터 문자 길이를 확인
print("<각 review 의 데이터 문자 길이> ")
train_length = train_data['review'].apply(len)
train_length.head()
```

```python
# 그래프에 대한 이미지 크기 선언
# figsize : (가로, 세로) 형태의 튜플로 입력
plt.figure(figsize=(12,5))
# 히스토그램 선언
# bins : 히스토그램 값에 대한 버킷 범위
# range : x축 값의 범위
# alpha : 그래프 색상 투명도
# color : 그래프 색상
# label : 그래프에 대한 라벨
plt.hist(train_length, bins=200, alpha=0.5, color='r', label='length')
plt.yscale('log', nonposy='clip')
# 그래프 제목
plt.title('Log-Histogram of length of review')
# 그래프 x축 라벨
plt.xlabel('Length of review')
# 그래프 y축 라벨
plt.ylabel('Number of review')
plt.show()
```

```python
print("<통계값 확인>")
print('리뷰 길이 최대값: {}'.format(np.max(train_length)))
print('리뷰 길이 최초값: {}'.format(np.min(train_length)))
print('리뷰 길이 평균값: {:.2f}'.format(np.mean(train_length)))
print('리뷰 길이 표준편차: {:.2f}'.format(np.std(train_length)))
print('리뷰 길이 중간값: {}'.format(np.median(train_length)))
# 사분위에 대한 경우는 0~100 스케일로 돼 있음
print('리뷰 길이 1사분위: {}'.format(np.percentile(train_length, 25)))
print('리뷰 길이 3사분위: {}'.format(np.percentile(train_length, 75)))
```

```python
print("<박스 플롯 생성>")
plt.figure(figsize=(12,5))
# 박스 플롯 생성
# 첫 번쨰 인자: 여러 분포에 대한 데이터 리스트 입력
# labels : 입력한 데이터에 대한 라벨
# showmeans: 평균값을 마크함

plt.boxplot(train_length, labels=['counts'], showmeans=True)
plt.show()
```

```python
print("<WordCloud로 단어 빈도수 확인>")

from wordcloud import WordCloud
cloud = WordCloud(width=600, height=400).generate(" ".join(train_data['review']))
plt.figure(figsize=(20,15))
plt.imshow(cloud)
plt.axis('off')
plt.show()
```
