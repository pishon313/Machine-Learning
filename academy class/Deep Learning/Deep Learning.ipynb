# 인공 뉴런

- sum(input(x) * w .. ) + bias
    - 입력과 가중치의 곱의 합 + bias
- activation func (활성화 함수)
- 딥러닝 학습이란, 가중치(기울기)와 절편(bias)을 구하는 과정
- 가중치와 바이아스를 랜덤으로 주지 말고, 고정해서 준 뒤, 직관적으로 살펴보자

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.arange(-1.0, 1.0, 0.2)
x2 = np.arange(-1.0, 1.0, 0.2)

Z = np.zeros((10,10))

# 가중치와 bias 값 고정 

w_x1 = 2.5
w_x2 = 3.0

bias = 0.1

for i in range(10):
    for j in range(10):
        u = x1[i] * w_x1 + x2[j] * w_x2 + bias
        y = 1 / (1 + np.exp(-u))
        Z[j][i] = y
        
plt.imshow(Z, "gray", vmin=0.0, vmax=1.0)
plt.colorbar()
plt.xlabel("x1")
plt.ylabel("y1")
plt.show()

# row: i, column = j
# 가중치와 bias를 고정시켜도 결과값이 다양하게 나올 수 있다. 
```

```python
X1 = np.arange(-1.0, 1.0, 0.2)
X2 = np.arange(-1.0, 1.0, 0.2)

Z = np.zeros((10,10))

# 가중치와 bias 값 고정 

W = np.array([2.5, 3.0])

bias = np.array([0.1])

for i in range(10):
    for j in range(10):
        list1 = [X1[i], X2[j]]
        X=np.array(list1)
        u=np.dot(X, W.T) + bias
        
        y =1 / (1 + np.exp(-u))
        Z[j][i] = y
            
            
plt.imshow(Z, "gray", vmin=0.0, vmax=1.0)
plt.colorbar()
plt.xlabel("x1")
plt.ylabel("y1")
plt.show()
```

4. 다중신경망
- 이제까지는 신경망의 열이 1개, 즉 출력값도 1개였다
- 이 출력값이 여러개가 될 수 있다. 출력 값이 여러 뉴런으로 이동 가능
- n개만큼 컬럼이 늘어날 수 있다. 
- 이떄, bias 역시 컬럼과 같은 n개가 필요하다 

- 입력층(Input layer)에서 은닉층(Hidden layer)으로 가는 길에 가중치(W) 와 바이아스를 정해줘서 은닉층(활성함수를 적용한 결과)가 나온다
- 은닉층은 맨 마지막으로 출력으로 보이는게 아니라 중간 계산 과정이기 떄문에 hidden 을 쓴다. 
- 계산중에는 사이 사이 존재하는 W와 B를 항상 신경써라 (w,b 을 안다는 것은 신경망이 어떻게 흘러가고 있는지 안다는 뜻이다))

### 다중 신경망

```python
# X, Y 가 feature
X = np.arange(-1.0, 1.0, 0.2)
Y = np.arange(-1.0, 1.0, 0.2)

Z = np.zeros((10,10))

# 입력에서 은닉으로 넘어가는 가중치 
w_im = np.array([[4.0, 4.0], [4.0, 4.0]]) # row, column 이 동일해야 하므로 (2,2) 행렬을 만들었다. 

# 은닉에서 아웃풋으로 갈때,
w_mo = np.array([[1.0], [-1.0]])

b_im = np.array([3.0, -3.0])
b_mo = np.array([0.1])

# 은닉층
def middle_layer(x,w,b):
    u = np.dot(x,w) + b
    return 1 / (1 + np.exp(-u))


# 출력층
def output_layer(x,w,b) :
    u = np.dot(x,w) + b
    return u 


for i in range(10):
    for j in range(10):
        inp = np.array([X[i], Y[j]])
        mid = middle_layer(inp, w_im, b_im)
        out = output_layer(mid, w_mo, b_mo)
        
        Z[j][i] = out[0] # 넘파이 단일값으로 나오기 때문에 아웃에 0으로 준다 
        
plt.imshow(Z, 'gray', vmin=0.0, vmax=1.0)
plt.colorbar()
plt.xlabel('Feature X')
plt.ylabel('Feature Y')
plt.show()
```

# 분류

```python
X = np.arange(-1.0, 1.0, 0.1)
Y = np.arange(-1.0, 1.0, 0.1)

w_im = np.array([[1.0, 2.0], [2.0, 3.0]])
w_mo = np.array([[-1.0, 1.0], [1.0, -1.0]])

# 출력이 2개 이므로, 바이너리 값도 2개가 된다. 
b_im = np.array([0.3, -0.3])
b_mo = np.array([0.4, 0.1])

def middle_layer(x, w, b):
    u = np.dot(x, w) + b
    return 1 / (1 + np.exp(-u))

def output_layer(x, w, b):
    u = np.dot(x, w) + b
    return np.exp(u) / np.sum(np.exp(u))

x_1 = []
y_1 = []
x_2 = []
y_2 = []

for i in range(20):
    for j in range(20):
        
        inp = np.array([X[i], Y[j]])
        mid = middle_layer(inp, w_im, b_im)
        out = output_layer(mid, w_mo, b_mo)
            
        # 따로 따로 저장
        if out[0] > out[1]:
            x_1.append(X[i])
            y_1.append(Y[j])
        else:
            x_2.append(X[i])
            y_2.append(Y[j])
                
plt.scatter(x_1, y_1, marker='+')
plt.scatter(x_2, y_2, marker='o')
plt.show()

# x,y 값을 대입하여 나온 값을 나눈 것 -> 분류 할 수 이다. 
# 가중치와 바이아스 를 고정했을 때, 분류하는 방법
```
