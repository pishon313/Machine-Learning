```python
import re
import json
from konlpy.tag import Okt
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
train_data.head()
```

```python
review_text = re.sub(r'[^가-힣ㄱ-ㅎㅏ-ㅣ\s]', '', train_data['document'][0])
print(review_text)
```

```python
okt = Okt()
review_text = okt.morphs(review_text, stem = True)
print(review_text)
```

```python
stop_words = set(['은', '는', '이', '가', '하', '아', '것', '들','의', '있', '되', '수', '보', '주', '등', '한'])
clean_review = [token for token in review_text if not token in stop_words]
print(clean_review)
```

```python
def preprocessing(review, okt, remove_stopwords = False, stop_words = []):
    review_text = re.sub(r'[^가-힣ㄱ-ㅎㅏ-ㅣ\s]', '', review)
    word_review = okt.morphs(review_text, stem = True)
    if remove_stopwords:
        word_review = [token for token in word_review if not token in stop_words]
        
    return word_review
```

```python
clean_train_review = []

for review in train_data['document']:
    if type(review) == str:
        clean_train_review.append(preprocessing(review, okt, remove_stopwords = True, 
                                               stop_words = stop_words))
    
    else:
        clean_train_review.append([])
        
print(clean_train_review[:4])
```

```python
test_data = pd.read_csv(DATA_IN_PATH + 'ratings_test.txt', delimiter = '\t', quoting = 3)

clean_test_review = []

for review in test_data['document']:
    if type(review) == str:
        clean_test_review.append(preprocessing(review, okt, remove_stopwords = True, 
                                               stop_words = stop_words))
    
    else:
        clean_test_review.append([])
        
print(clean_test_review[:4])
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(clean_train_review)
train_sequences = tokenizer.texts_to_sequences(clean_train_review)
test_sequences = tokenizer.texts_to_sequences(clean_test_review)

word_vocab = tokenizer.word_index
```

```python
print(word_vocab['더빙'])
```

```python
MAX_SEQUENCE_LENGTH = 8

train_inputs = pad_sequences(train_sequences, maxlen = MAX_SEQUENCE_LENGTH, padding = 'post')
train_labels = np.array(train_data['label'])

test_inputs = pad_sequences(test_sequences, maxlen = MAX_SEQUENCE_LENGTH, padding = 'post')
test_labels = np.array(test_data['label'])
```

```python
DATA_IN_PATH = './data_in/'
TRAIN_INPUT_DATA = 'nsmc_train_input.npy'
TRAIN_LABEL_DATA = 'nsmc_train_label.npy'
TEST_INPUT_DATA = 'nsmc_test_input.npy'
TEST_LABEL_DATA = 'nsmc_test_label.npy'
DATA_CONFIGS = 'nsmc_data_configs.json'

data_configs = {}
data_configs['vocab'] = word_vocab
data_configs['vocab_size'] = len(word_vocab)

import os
if not os.path.exists(DATA_IN_PATH):
    os.makedirs(DATA_IN_PATH)
    
np.save(open(DATA_IN_PATH + TRAIN_INPUT_DATA, 'wb'), train_inputs)
np.save(open(DATA_IN_PATH + TRAIN_LABEL_DATA, 'wb'), train_labels)

np.save(open(DATA_IN_PATH + TEST_INPUT_DATA, 'wb'), test_inputs)
np.save(open(DATA_IN_PATH + TEST_LABEL_DATA, 'wb'), test_labels)

json.dump(data_configs, open(DATA_IN_PATH + DATA_CONFIGS, 'w'), ensure_ascii = False)
```
