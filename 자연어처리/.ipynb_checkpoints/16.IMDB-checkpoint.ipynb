```python
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Embedding, LSTM
from tensorflow.keras.layers import Conv1D, MaxPooling1D
from tensorflow.keras.datasets import imdb
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test ) = imdb.load_data(num_words=5000)
```

```python
x_train = sequence.pad_sequences(x_train, maxlen=100)
x_test = sequence.pad_sequences(x_test, maxlen=100)
```

```python
model = Sequential()
model.add(Embedding(5000,100))
model.add(Dropout(0.5))
model.add(Conv1D(64,5, padding='valid', activation='relu', strides=1))
# strides =1 : 한 칸씩 이동한다. 
model.add(MaxPooling1D(pool_size=4))
# 위는 1차원이기 떄문에 4칸이 된다. 
model.add(LSTM(55))
#LSTM 통과
model.add(Dense(1))
# 결과는 1개, 
model.add(Activation('sigmoid'))
# activation 클래스를 사용하는 것

          
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# 모델 훈련
hist = model.fit(x_train, y_train, batch_size=100, epochs=10, validation_data=(x_test, y_test))
```

```python
print("Test Accuracy: % .4f" % (model.evaluate(x_test, y_test)[1]))
```

```python
y_vloss = hist.history['val_loss']
y_loss = hist.history['loss']

x_len = np.arange(len(y_loss))
plt.plot(x_len, y_vloss, marker='.', c='red', label='Test_losss')
plt.plot(x_len, y_loss, marker='.', c='blue', label='Trainset_loss')
plt.legend()
plt.grid()
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```

결과를 보면, 첫번쨰 두번쨰 세번쨰까지는 테스트로서 계속 올라간다 
이에 반해 훈련셋은 떨어진다
